```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Data from your image (2000 images total)
cm = np.array([[996, 4],
               [7, 993]])

total = 2000
accuracy = (996 + 993) / total * 100

# Create figure matching your exact style
fig, ax = plt.subplots(figsize=(8, 6))

# Heatmap with exact color scheme and styling
sns.heatmap(cm, annot=True, fmt='d', cmap='hot_r',
            xticklabels=['Fake', 'Real'],
            yticklabels=['Fake', 'Real'],
            ax=ax,
            cbar_kws={'label': '', 'shrink': 0.8},
            linewidths=1, linecolor='white',
            square=True)

# Title and labels matching your image
ax.set_title('Confusion Matrix\nAccuracy: 99.45%', pad=20)
ax.set_xlabel('Predicted', fontsize=12, labelpad=15)
ax.set_ylabel('Actual', fontsize=12, labelpad=15)

# Rotate x-axis labels to match your image angle
plt.setp(ax.get_xticklabels(), rotation=0)
plt.setp(ax.get_yticklabels(), rotation=0)

# Customize colorbar ticks to match your image
cbar = ax.collections[0].colorbar
cbar.set_ticks([0, 200, 400, 600, 800])
cbar.set_ticklabels(['0', '200', '400', '600', '800'])

# Adjust layout
plt.tight_layout()
plt.savefig("confusion_matrix.png", dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none')
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Dataset: 140002 images, split evenly
total = 140000
fake_total = total // 2       # 70001
real_total = total - fake_total  # 70001

# Perfect classification — no misclassifications
cm = np.array([[fake_total, 0],
               [0, real_total]])

accuracy = (cm[0][0] + cm[1][1]) / total * 100

fig, ax = plt.subplots(figsize=(7, 6))

sns.heatmap(cm, annot=True, fmt='d', cmap='hot_r',
            xticklabels=['Fake', 'Real'],
            yticklabels=['Fake', 'Real'],
            ax=ax, linewidths=0.5)

ax.set_title(f'Confusion Matrix\nAccuracy: {accuracy:.2f}%', fontsize=13, pad=12)
ax.set_xlabel('Predicted', fontsize=11)
ax.set_ylabel('Actual', fontsize=11)

plt.tight_layout()
plt.savefig("confusion_matrix_140002.png", dpi=300, bbox_inches='tight')
plt.show()
```
