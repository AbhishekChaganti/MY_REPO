```
from google_play_scraper import reviews, Sort
import pandas as pd
from datetime import datetime

# App package name for Zepto
app_package = "com.zeptoconsumerapp"

# Define date range (Change year if needed)
start_date = datetime(2026, 1, 1)
end_date = datetime(2026, 1, 31, 23, 59, 59)

all_reviews = []
continuation_token = None

while True:
    result, continuation_token = reviews(
        app_package,
        lang='en',
        country='in',
        sort=Sort.NEWEST,
        count=200,
        continuation_token=continuation_token
    )

    if not result:
        break

    for review in result:
        review_date = review['at']
        if start_date <= review_date <= end_date:
            all_reviews.append(review)
        elif review_date < start_date:
            break

    if review_date < start_date:
        break

# Convert to DataFrame
df = pd.DataFrame(all_reviews)

# Keep important columns
df = df[['reviewId', 'userName', 'score', 'at', 'content', 'thumbsUpCount']]

# Save to CSV
df.to_csv("zepto_reviews_january_2026.csv", index=False)

print("CSV file created successfully!")
print("Total reviews collected:", len(df))
```
